1. Create a numpy array on size 2 x 2

```python
arr_2x2 = np.array([[1,2],[3,4]])

print("2x2 Array:\n", arr_2x2)
```

2. Create a random numpy array of size 3 x 4

```python
random_arr_3x4 = np.random.rand(3,4)

print("3x4 Random Array:\n", random_arr_3x4)
```

3. Create a numpy array of ones and a numpy array of zeroses

```python
ones_array = np.ones((2, 3))
zeros_array = np.zeros((2, 3))

print("Ones Array:\n", ones_array)
print("\nZeros Array:\n", zeros_array)
```

4. Create a random numpy array of size 6x3 and transpose it

```python
random_arr_6x3 = np.random.rand(6, 3)
transposed_arr = np.transpose(random_arr_6x3)

print("6x3 Random Array:\n", random_arr_6x3)
print("\nTransposed Array:\n", transposed_arr)
```

5. Create two random numpy array of size n x p and p x m and then compute the dot product

```python
import numpy as np


n = 2
p = 3
m = 4

array_np = np.random.rand(n, p)
array_pm = np.random.rand(p, m)

dot_product = np.dot(array_np, array_pm)

print("Array np (n x p):\n", array_np)
print("\nArray pm (p x m):\n", array_pm)
print("\nDot Product Result (n x m):\n", dot_product)
```
